# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Importing dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

# Training LR model on whole dataset


```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()   #object created
lin_regressor.fit(x,y)   #training the object on the model
# y_pred = lin_regressor.predict(x)
```

# Training PR model on whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor = PolynomialFeatures(degree = 4)  #new object
x_poly = poly_regressor.fit_transform(x)   # training the model

lin_regressor_2 = LinearRegression()  #new object 2
lin_regressor_2.fit(x_poly,y)  # training model on new matrix of features
```

# Visualizing LR results


```python
plt.scatter(x,y,color='red')
plt.plot(x, lin_regressor.predict(x),color='blue')
plt.title('Truth vs Bluff (LR)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# Visualizing PR results

```python
plt.scatter(x, y, color= 'red')
plt.plot(x,lin_regressor_2.predict(x_poly), color= 'blue')
plt.title('Truth vs Bluff (PR)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# Visualizing PR result(for higher resoltion and smoother curve)

```python
x_grid = np.arange(min(x),max(x),0.1)  # increasing the density of position levels 1, 1.1 ...
x_grid = x_grid.reshape((len(x_grid),1))
y_grid = poly_regressor.fit_transform(x_grid)


plt.scatter(x, y, color= 'red')
plt.plot(x_grid,lin_regressor_2.predict(y_grid), color= 'blue')
plt.title('Truth vs Bluff (PR smooth)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# Predicting salary for 6.5 level using LR

```python
print(lin_regressor.predict([[6.5]]))   #prediction way more than expected 160k dollars
```

# Predicting salary for 6.5 level using PR

```python
print(lin_regressor_2.predict(poly_regressor.fit_transform([[6.5]])))  #close to 160k dollar
```
